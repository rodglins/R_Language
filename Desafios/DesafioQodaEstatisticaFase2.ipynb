### **31) Vendas**

```
#34 Vendas
# Distribuição de frequência:
vendasJulho <- c(2114, 2468, 7119, 1876, 4105, 3183, 1932, 1355, 4278, 1030, 2000, 1077, 5835, 1512, 1697, 2478, 3981, 1643, 1858, 1500, 4608, 1000)
tabela <- data.frame(t(table(vendasJulho)))[,-1]
tabela$vendasJulho <- as.numeric(levels(tabela$vendasJulho))
tabela <- tabela %>% 
  mutate(Fr = 100*Freq/sum(Freq),
         Fac = cumsum(Freq),
         Xi.Fi = vendasJulho*Freq)
tabela
```
